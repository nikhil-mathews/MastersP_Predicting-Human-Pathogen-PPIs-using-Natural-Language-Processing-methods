```python
def to_df(name):
    df = pd.read_csv("C:/Users/nik00/py/proj/"+name+".txt", sep="\t", header=None)
    df.columns = df.iloc[0,:]
    df = df.drop([0], axis=0).reset_index(drop=True)
    return df
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df_train1 = to_df('HVtrain_set_group1')
df_train2 = to_df('HVtrain_set_group2')
df_train3 = to_df('HVtrain_set_group3')

df_test1 = to_df('HVindependent_test_group1')
df_test2 = to_df('HVindependent_test_group2')
df_test3 = to_df('HVindependent_test_group3')

data1 = pd.concat([df_train1,df_test1])
data2 = pd.concat([df_train2,df_test2])
data3 = pd.concat([df_train3,df_test3])

all_trains = pd.concat([df_train1,df_train2,df_train3]).drop_duplicates().reset_index(drop=True)
all_tests = pd.concat([df_test1,df_test2,df_test3]).drop_duplicates().reset_index(drop=True)
all_HV_data = pd.concat([all_trains,all_tests]).drop_duplicates().reset_index(drop=True)
```

```python

nans = []
all_data = pd.read_pickle("C:/Users/nik00/py/proj/all_data.pkl")
all_data.head()


for idd in all_data.index:
    if type(all_data['Sequence'].loc[idd]) is float:
        #print(idd)
        nans.append(idd)
        
len(nans)

df = all_HV_data.copy()



train2_ids = list(df.Human_ID)+list(df.Virus_ID)
len(train2_ids)

for idd in train2_ids:
    if idd not in all_data.index:
        nans.append(idd)
len(nans)
```

50,000 datapoints lost after web scraping

```python
df = df[~df['Human_ID'].isin(nans)]
df = df[~df['Virus_ID'].isin(nans)]
print(df.shape)
df["Human"] = [list(all_data.loc[df.iloc[row]["Human_ID"]]["Sequence"]) for row in range(df.shape[0])]
df["Virus"] = [list(all_data.loc[df.iloc[row]["Virus_ID"]]["Sequence"]) for row in range(df.shape[0])]
df
```
